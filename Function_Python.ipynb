# Function of printing all list in one **line**

```python
tools = ["hammer", "screw driver", "plass", "saw", "knife", "drill"]
def cal_tools(list):
     for items in list:
          print(items, end=" ")
          # print(items, end="\n")
cal_tools(tools)
```
